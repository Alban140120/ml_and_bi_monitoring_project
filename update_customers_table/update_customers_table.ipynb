```python
%sql
-- Création de la table temporaire pour insérer les données depuis le fichier CSV
CREATE TABLE IF NOT EXISTS amavla_dev_central_adbw.default.tmp_nbocustomers (
    customer_id VARCHAR(10),
    first_name VARCHAR(20),
    last_name VARCHAR(20),
    gender BOOLEAN,
    nationality VARCHAR(20),
    birth_date DATE,
    street_address VARCHAR(50),
    city VARCHAR(50),
    postal_code VARCHAR(10),
    country VARCHAR(20),
    phone_number VARCHAR(20),
    phone_type BOOLEAN,
    email_address VARCHAR(50)
);
```

```python
%sql
-- Charger le CSV dans cette table
COPY INTO tmp_nbocustomers(customer_id, first_name, last_name, gender, nationality, birth_date, street_address, city, postal_code, country, phone_number, email_address)
FROM 'C:/Users/alban.daniel_amaris/OneDrive - Mantu Group/Documents/Projet NBO/fichier.csv'
FILEFORMAT = CSV;
```

```python
%sql
-- Mise à jour des customers existants
UPDATE nbocustomers AS n
SET
    first_name = t.first_name,
    last_name = t.last_name,
    gender = t.gender,
    nationality = t.nationality,
    birth_date = t.birth_date,
    street_address = t.street_address,
    city = t.city,
    postal_code = t.postal_code,
    country = t.country,
    phone_number = t.phone_number,
    phone_type = t.phone_type,
    email_address = t.email_address
FROM tmp_nbocustomers AS t
WHERE n.customer_id = t.customer_id;
```

```python
%sql
INSERT INTO nbocustomers (customer_id, first_name, last_name, gender, nationality, birth_date, street_address, city, postal_code, country, phone_number, phone_type, email_address)
SELECT t.customer_id
      , t.first_name
      , t.last_name
      , t.gender
      , t.nationality
      , t.birth_date
      , t.street_address
      , t.city
      , t.postal_code
      , t.country
      , t.phone_number
      , t.phone_type
      , t.email_address
FROM tmp_nbocustomers AS t
WHERE t.customer_id NOT IN 
      (SELECT customer_id FROM nbocustomers)
;
```
